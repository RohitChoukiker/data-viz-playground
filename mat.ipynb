```python
x=[1, 2, 0, 4, 5]
y=[5, 4, 3, 2, 1]
import matplotlib.pyplot as plt
plt.plot(x,y, color="black", linestyle="--", marker="o", markerfacecolor="red", markersize=6)
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.grid(True)
plt.title('Simple Line Plot')
plt.legend(['Learning 1'])
plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9,top=0.9,wspace=None,hspace=None)  

plt.show()
```

```python
data = [1, 2, 3 ,3,1,2,4 ,5, 6, 7, 8, 9, 10]
import matplotlib.pyplot as plt
plt.hist(data, bins=    5,color='black', alpha=0.7, edgecolor='green')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Data')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['red', 'green', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)

# Plot
plt.pie(
    sizes,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    pctdistance=0.85,
    shadow=True,
    startangle=90
)

# Make it a perfect circle
plt.axis('equal')

# Title
plt.title('Pie Chart Example')

# Show chart
plt.show()
```
