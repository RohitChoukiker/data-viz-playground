```python
x=[1, 2, 0, 4, 5]
y=[5, 4, 3, 2, 1]
import matplotlib.pyplot as plt
plt.plot(x,y, color="black", linestyle="--", marker="o", markerfacecolor="red", markersize=6)
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.grid(True)
plt.title('Simple Line Plot')

plt.show()
```
